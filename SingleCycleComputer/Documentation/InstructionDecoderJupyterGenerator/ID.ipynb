```python
d = data.drop(['name','name2','opcode'],axis=1)
for var in d[:]:
    print(var)
```

```python
#inner case
for i,op in enumerate(opcode):
    print('`'+name[i]+':'+ ' \tbegin')
    for k in d:
        bin = len(str(d[k][i]))
        print('\t'+k+' <= '+str(bin)+'\'b'+str(d[k][i])+';')
        if k=="op":
            print('\tend')
```
